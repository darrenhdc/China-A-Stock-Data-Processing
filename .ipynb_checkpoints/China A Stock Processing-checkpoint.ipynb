```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

def convert_str_to_datetime(date: str)-> datetime:
    '''
    convert '2021-3-4' to datetime format
    '''
    date1 = date.split('-')
    stamp = datetime(int(date1[0]), int(date1[1]), int(date1[2]))
    return stamp

def format_date_to_8_digits_str(date: str) -> str:
    '''
    convert '2021-3-4' to 8-digits str format '20210304'
    '''
    stamp = convert_str_to_datetime(date)
    return stamp.strftime('%Y%m%d')
    
date = '2021-3-4'
format_date_to_8_digits_str(date)
```

```python
class StockData:
    def __init__(self, path):
        '''
        Initialize StockData Object
        path: data folder path
        df: pandas dataframe to store stock data
        '''
        self.path = path # folder path
        self.frames = dict() # dictionary to store chosen stock data, key is symbol, value is a dataframe

    def read(self, symbols: list):
        '''
        specify the symbols of stocks, 
        return a single dataframe containing the chosen symbols
        '''
        frames = [pd.read_csv('{}/{}.csv'.format(self.path, symbol)) for symbol in symbols]
        
        # reformat field 'trade_dt', '2021-3-4' to '20210304'
        frames_reformat = []
        for frame in frames:
            frame['trade_dt'] = frame['trade_dt'].apply(lambda x: format_date_to_8_digits_str(x))
            frames_reformat.append(frame)
            
        self.frames = dict(zip(symbols, frames_reformat))
        
        
    def get_data_by_symbol(self, symbol: str, start_date: str, end_date: str) -> pd.DataFrame:
        '''specify a symbol, output the trading info between a range of dates '''
        try:
            df = self.frames[symbol]
        except:
            print('Error: the symbol is not loaded by StockData.read(), please reload and try again.')
        start_date = format_date_to_8_digits_str(start_date)
        end_date = format_date_to_8_digits_str(end_date)      
        ret = df.loc[(df['trade_dt'] >= start_date) & (df['trade_dt'] <= end_date)];
        ret = ret[['trade_dt', 's_dq_open', 's_dq_high', 's_dq_low', 's_dq_close', 's_dq_volume']]
        return ret

    def get_data_by_date(self, adate: str, symbols: list) -> pd.DataFrame:
        '''specify a certain date, output trading info of chosen stocks'''
        adate = format_date_to_8_digits_str(adate)
        df_list = []
        for symbol in symbols:
            if symbol not in self.frames.keys():
                print("symbol {} is not loaded by StockData.read()")
                break
            df = self.frames[symbol]  
            df = df.loc[df['trade_dt'] == adate]           
            df['symbol'] = symbol
            df_list.append(df)
        ret = pd.concat(df_list)
        ret = ret[['symbol', 'trade_dt', 's_dq_open', 's_dq_high', 's_dq_low', 's_dq_close', 's_dq_volume']]
        return ret
        
    def get_data_by_filed(self, field, symbols):
        '''not clear for the question'''
        pass
    
    def format_date(self, symbol: str)-> pd.DataFrame:
        ''' Convert given symbol's data index to pandas datetime object '''
        try:
            df = self.frames[symbol]
        except:
            print('Error: the symbol is not loaded by StockData.read(), please reload and try again.')
        df['trade_dt'] = df['trade_dt'].apply(lambda x: pd.to_datetime(x, format = '%Y%m%d'))
        df = df.set_index('trade_dt')
        df = df[[ 's_dq_open', 's_dq_high', 's_dq_low', 's_dq_close', 's_dq_volume']]
        return df
    
    def plot(self, symbol: str, field):
        ''' plot the scatter of chosen field, x-axis is date, y-axis is value '''
        try:
            df = self.frames[symbol]
        except:
            print('Error: the symbol is not loaded by StockData.read(), please reload and try again.')
        df['trade_dt'] = df['trade_dt'].apply(lambda x: pd.to_datetime(x, format = '%Y%m%d'))
        x = list(df['trade_dt'])
        y = list(df[field])
        plt.figure()
        plt.plot(x, y)        
    
    def adjust_data(self, symbol: str) -> pd.DataFrame:
        '''compute back/forward adjusted price'''
        df = stock.frames[symbol]
        df['AF'] = np.nan
        df['AF'][0] = np.float64(1)
        for i in range(1,len(df)):
            last_day_AF = df['AF'][i-1]
            last_day_close = df['s_dq_close'][i-1]
            today_close = df['s_dq_close'][i]
            today_preclose = df['s_dq_preclose'][i]
            today_AF = last_day_close / today_preclose * last_day_AF
            print(last_day_close / today_preclose)
            df['AF'][i] = today_AF
            df['back_adjusted_price'] = today_close * today_AF
            df['forward_adjusted_price'] = today_close * today_AF * df['AF'][len(df)-1]
        return df
    
    def resample(self, symbol: str, freq: int):
        '''given a frequency and resample the DataFrame according to it'''
        try:
            df = self.frames[symbol]
        except:
            print('Error: the symbol is not loaded by StockData.read(), please reload and try again.')
        print("Original total number:", len(df))
        indices = [x for x in range(0, len(df), freq)]
        df = df.iloc[indices]
        df = df[[ 's_dq_open', 's_dq_high', 's_dq_low', 's_dq_close', 's_dq_volume']]
        print("Total number after resample:", len(df))
        return df
```
